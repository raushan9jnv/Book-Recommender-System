## Popularity based Recomender System

We will display top 50 books with Highest average Ratings but we will consider only those books which have got a  minimum of 250 vote.

```python
# Merge ratings and books dataset
rating_with_names = ratings.merge(books, on ="ISBN")

#shape of merge data
rating_with_names.shape
```

 here the ratings number become 1031136 from 1149780 reason is some rating's books are not present in book dataset so some rows decreases.

```python
# this dataset there are multiple books whose name are same  but there isbn are same means harry potter books are avaiable with three different isbn number.
num_rating_df = rating_with_names.groupby("Book-Title").count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating': 'num_ratings'}, inplace=True)
num_rating_df
```

```python
#averge rating
avg_rating_df = rating_with_names.groupby("Book-Title").mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns = {'Book-Rating': 'avg_ratings'}, inplace=True)
avg_rating_df
```

```python
#merge num_rating_df and avg_rating_df
popular_df = num_rating_df.merge(avg_rating_df, on = 'Book-Title')
popular_df
```

```python
# ratings greater or equal to 250 and select top 50
popular_df = popular_df[popular_df['num_ratings'] >= 250].sort_values('avg_ratings', ascending = False).head(50)
popular_df
```

```python
popular_df = popular_df.merge(books, on= 'Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author','Image-URL-M', 'num_ratings', 'avg_ratings']]
```

```python
# these are our top 50 books and there details based on Popularity based recommender system using average rating formula
popular_df
```

# Collaborative filtering based recommender system

after working on this data, I relise, in this we are not considering all users  and books in model building as it was not giving accurate predicitions.
- So after error and trial method (EXPERIMENT) i come to conclusion that we are considering only those user who have given minimum 200 reviews as they are experienced and we considering only those books which have minimum 50 reviws given by people. so by this we negelect unpopular book which are known by very less people. to make our model/system more intelligent.

USER > 200 REVIEW GIVEN

BOOKS, REVIEW BY >= 50 PEOPLE

```python
#filtering based on user
x = rating_with_names.groupby('User-ID').count()['Book-Rating'] > 200
user_200 = x[x].index
user_200
```

```python
filtered_rating = rating_with_names[rating_with_names['User-ID'].isin(user_200)]
```

```python
#filtering based on books

y =filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
final_ratings.drop_duplicates()
```

### pivot table

```python
#these are the people who read more than 200 books and books are minimum reviewed by 50 peoples. represented by pivot table
pt = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
pt
```

there is 706 famous books and 810 trusted users

```python
pt.fillna(0, inplace = True)
pt
```

Now calculate  eucliden distances

```python
from  sklearn.metrics.pairwise import cosine_similarity
```

```python
#calculate cosine similarity of all rows with oother rows, it will give 2D numpy array
similarity_scores = cosine_similarity(pt)
similarity_scores
```

```python
similarity_scores.shape
```

```python
similarity_scores[0]
```

```python
def recommend(book_name):
  #index fetch
  index = np.where(pt.index == book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_scores[index])),key = lambda x:x[1], reverse = True)[1:6]

 # for print only books name

  # for i in similar_items:
  #   print(pt.index[i[0]])

# print books author image url

  data =[]
  for i in similar_items:
    item = []
    temp_df = books[books['Book-Title'] == pt.index[i[0]]]
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values)) # don't use APPEND with LIST use EXTEND because list become 2D array after append.
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

    data.append(item)

  return data
```

```python
recommend('1984')
```
